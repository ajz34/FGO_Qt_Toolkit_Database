```python
def semi_automatic_xml_export(identity_title_name):
    for identity, title_name in identity_title_name:
        time.sleep(5)
        identity_replaced = "{:03d}".format(identity)
        title_name_replaced = title_name.replace("・", "·")
        try:
            mooncell_raw = \
                requests.get("https://fate.wiki/index.php", params={"title": title_name_replaced, "action": "edit"})
            mooncell_html = bs4.BeautifulSoup(mooncell_raw.content, "lxml")
            mooncell_text = mooncell_html.find_all("textarea")[0].get_text()
            mooncell_text = mooncell_text.split('\n')
            with open("requested_file/" + identity_replaced + ".txt", "w", encoding="utf-8") as file:
                for string in mooncell_text:
                    file.write(string)
                    file.write("\n")
            print(identity_replaced)
        except:
            print("Error: " + identity_replaced)
```

```python
# 定义字符分段
class mooncell_assist:
    
    def __init__(self, text):
        self.text = text   # Text of the current prasing line
        self.list = None   # If parsed successfully, list data is stored here
        self.dict = None   # If parsed successfully, dict data is stored here
        self.succ = False  # Indication of the current parse
    
    def parse(self):
        # 1. let the text to be single line
        if (type(self.text) == str):
            self.text = [self.text]
        text = ""
        for tempstr in self.text:
            text += tempstr.replace("\n", "")
        self.text = text
        # 2. check if the leftmost `{{` is paired with rightmost `}}`, then remove them
        # then append `|` at last, indicate the last cell
        if (text[:2] == "{{") and (text[-2:] == "}}"):
            text = text[2:-2] + "|"
        # 3. split to list by `|` and `{{` and `}}` character by character
        lst = []
        dct = {}
        dct_flag = False
        braket_flag = False
        success_flag = True
        item = ""
        val = ""
        braket_string = ""
        for c in text:
            # print(c)
            if braket_flag == True:
                # count until the last `}` exists
                braket_string += c
                if c == "{":
                    braket_count += 1
                elif c == "}":
                    braket_count -= 1
                if braket_count == 0:
                    braket_flag = False
                    if not dct_flag:
                        # return when list mode
                        item_tmp = mooncell_assist(braket_string)
                        item_tmp.parse()
                        item.append(item_tmp)
                        braket_string = ""
                        continue
                    else:
                        # return when dict mode
                        val_tmp = mooncell_assist(braket_string)
                        val_tmp.parse()
                        val.append(val_tmp)
                        braket_string = ""
                        continue
                else:
                    continue
            if (c not in ["|", "{", "}", "="]):
                # parse as trival character
                # check if item or val is string mode; if not, we treat them as a new item
                if not dct_flag:
                    if (type(item) == str):
                        item += c
                    else:
                        # end of this cell
                        lst.append(item)
                        item = ""
                        val = ""
                        item += c
                else:
                    if (type(val) == str):
                        val += c
                    else:
                        # end of this cell
                        dct[item] = val
                        dct_flag = False
                        item = ""
                        val = ""
                        item += c
                continue
            elif (c == "{"):
                # read the information list, then set it to type mooncell_assist, recursively
                # at the same time, val or item formulation should be changed to list, containing mooncell_assist type of items
                braket_count = 1
                braket_string = c
                braket_flag = True
                if dct_flag:
                    if (type(val) == str) and (val != ""):
                        # =foo{{bar}} string is not supported
                        print("Possibly Error! Type 2")
                        success_flag = False
                    elif (type(val) == str):
                        # first time to have {{}}, creat list
                        val = []
                else:
                    if (type(item) == str) and (item != ""):
                        # return the last item, and treat as the first item of the list
                        lst.append(item)
                        item = []
                        val = ""
                    elif (type(item) == str):
                        # first time to have {{}}, creat list
                        item = []
                        val = ""
                continue
            elif (c == "|"):
                # end of this cell, append item to list or dict
                if not dct_flag:
                    ## # check if item is empty
                    ## if item != "":
                    ##     lst.append(item)
                    lst.append(item)
                else:
                    dct[item] = val
                    dct_flag = False
                item = ""
                val = ""
                continue
            elif (c == "="):
                # set flag to dict, start to read characters as val
                dct_flag = True
                continue
        if braket_flag == True:
            # braket not compared
            print("Possibly Error! Type 1")
            success_flag = False
        self.list = lst
        self.dict = dct
        # end
        return self
```

```python
# 1. 基础信息定义
def parse_servant_dict_basic(servant_dict, mooncell_text):
    parse_read_flag = -1
    parse_read_string = ""
    parse_read_count = 0
    for line in mooncell_text:
        if line == "": continue
        if line == "==基础数值==":
            parse_read_flag = 0
            servant_dict["basic"] = {}
            continue
        if (line[2:6] == "基础数值") and (parse_read_flag == 0):
            parse_read_flag = 1
        if parse_read_flag >= 1:
            for c in line:
                if c == "{":
                    parse_read_count += 1
                    parse_read_flag = 2
                elif c == "}":
                    parse_read_count -= 1
                if parse_read_flag == 2:
                    parse_read_string += c
                if parse_read_count == 0 and parse_read_flag == 2:
                    parse_read_flag = 3
                    break
            if parse_read_flag == 3:
                break
    servant_assist = mooncell_assist(parse_read_string)
    servant_assist.parse()
    servant_dict["basic"]["rarity"] = int(servant_assist.dict["稀有度"])
    servant_dict["basic"]["id"] = int(servant_assist.dict["序号"])
    servant_dict["basic"]["name_sc"] = servant_assist.dict["姓名"]
    servant_dict["basic"]["name_jp"] = servant_assist.dict["日文名"]
    servant_dict["basic"]["name_en"] = servant_assist.dict["英文名"]
    servant_dict["basic"]["class"] = servant_assist.dict["职阶"]
    servant_dict["basic"]["alignments"] = servant_assist.dict["属性"].split("·")
    servant_dict["basic"]["gender"] = servant_assist.dict["性别"]
    servant_dict["basic"]["attribute"] = servant_assist.dict["阵营"]
    servant_dict["basic"]["atk_base"] = int(servant_assist.dict["基础ATK"])
    servant_dict["basic"]["hp_base"] = int(servant_assist.dict["基础HP"])
    servant_dict["basic"]["atk_4"] = int(servant_assist.dict["灵基4ATK"])
    servant_dict["basic"]["hp_4"] = int(servant_assist.dict["灵基4HP"])
    if servant_dict["basic"]["id"] != 1:
        servant_dict["basic"]["atk_90"] = int(servant_assist.dict["90级ATK"])
        servant_dict["basic"]["hp_90"] = int(servant_assist.dict["90级HP"])
        servant_dict["basic"]["atk_100"] = int(servant_assist.dict["100级ATK"])
        servant_dict["basic"]["hp_100"] = int(servant_assist.dict["100级HP"])
        servant_dict["basic"]["atk_max"] = int(servant_assist.dict["MAX强化ATK"])
        servant_dict["basic"]["hp_max"] = int(servant_assist.dict["MAX强化HP"])
    else:
        servant_dict["basic"]["atk_90"] = 0
        servant_dict["basic"]["hp_90"] = 0
        servant_dict["basic"]["atk_100"] = 0
        servant_dict["basic"]["hp_100"] = 0
        servant_dict["basic"]["atk_max"] = 0
        servant_dict["basic"]["hp_max"] = 0
    servant_dict["basic"]["card"] = "".join([ servant_assist.dict[i][0] for i in ["第一张卡", "第二张卡", "第三张卡", "第四张卡", "第五张卡"] ])
    # servant_dict["basic"]["characteristic"] = servant_assist.dict["特性"].replace("、", "，").replace(",", "，").split("，")
    # servant_dict["basic"]["characteristic"] = [ i for i in servant_dict["basic"]["characteristic"] if i != ""]
    servant_dict["basic"]["characteristic"] = []
    for i in servant_assist.dict:
        if i[:2] == "特性":
            servant_dict["basic"]["characteristic"].append(servant_assist.dict[i])
    if (type(servant_assist.dict["昵称"]) == str):
        servant_dict["basic"]["nickname"] = servant_assist.dict["昵称"].replace("、", "，").replace(",", "，").split("，")
        servant_dict["basic"]["nickname"] = [ i for i in servant_dict["basic"]["nickname"] if i != "" ]
    else:
        raise ValueError("Have Something Dirty!")
    return servant_dict
```

```python
# 2. 宝具定义
def parse_servant_dict_nobel(servant_dict, mooncell_text):
    parse_read_flag = -1
    parse_read_string = ""
    parse_read_count = 0
    servant_dict_noble = {}
    for line in mooncell_text:
        if line == "": continue
        if line == "==宝具==":
            parse_read_flag = 0
            servant_dict["nobel"] = {}
            servant_dict["nobel"]["revealable"] = False
            continue
        if parse_read_flag == 0:
            parse_read_flag = 1
            continue
        if parse_read_flag == 1:
            if line[0] != "{":
                servant_dict["nobel"]["upgradable"] = True
                parse_read_flag = 2
                continue
            else:
                servant_dict["nobel"]["upgradable"] = False
                parse_read_flag = 3
        if (parse_read_flag == 2) and (line[:3] == "强化后"):
            parse_read_flag = 3
        if (parse_read_flag == 2) and (line[:5] == "真名解放后"):
            parse_read_flag = 4
            servant_dict["nobel"]["revealable"] = True
            servant_dict["nobel"]["upgradable"] = False
        if (parse_read_flag == 3) and (line[2:4] == "宝具"):
            parse_read_flag = 4
        if parse_read_flag >= 4:
            for c in line:
                if c == "{":
                    parse_read_flag = 5
                    parse_read_count += 1
                elif c == "}":
                    parse_read_count -= 1
                if parse_read_flag == 5:
                    parse_read_string += c
                if parse_read_flag == 5 and parse_read_count == 0:
                    parse_read_flag = 6
            if parse_read_flag == 6:
                break
    servant_assist = mooncell_assist(parse_read_string)
    servant_assist.parse()
    servant_dict["nobel"]["card"] = servant_assist.list[1][0]
    servant_dict["nobel"]["level"] = servant_assist.list[2]
    servant_dict["nobel"]["name_sc"] = servant_assist.list[6]
    servant_dict["nobel"]["name_jp"] = servant_assist.list[4]
    servant_dict["nobel"]["name_en"] = servant_assist.list[5]
    return servant_dict
```

```python
# 3. 定义技能
def parse_servant_dict_skill(servant_dict, mooncell_text, skill_count):
    parse_read_flag = 0
    parse_read_string = ""
    parse_read_count = 0
    for line in mooncell_text:
        if line == "": continue
        if line[:6] == "'''技能" + str(skill_count):
            parse_read_flag = 1
            servant_dict["skill_" + str(skill_count)] = {}  # 这种模式可能允许三技能不存在时的情况
            continue
        if parse_read_flag == 1:
            if line[2:6] != "持有技能":
                servant_dict["skill_" + str(skill_count)]["upgradable"] = True
                parse_read_flag = 3
                continue
            else:
                servant_dict["skill_" + str(skill_count)]["upgradable"] = False
                parse_read_flag = 4
        if parse_read_flag == 3:
            if line[2:6] == "持有技能":
                parse_read_flag = 4
        if parse_read_flag >= 4:
            for c in line:
                if c == "{":
                    parse_read_flag = 5
                    parse_read_count += 1
                elif c == "}":
                    parse_read_count -= 1
                if parse_read_flag == 5:
                    parse_read_string += c
                if parse_read_count == 0 and parse_read_flag == 5:
                    parse_read_flag = 6
                    break
            if parse_read_flag == 6:
                break
    servant_assist = mooncell_assist(parse_read_string)
    servant_assist.parse()
    servant_dict["skill_" + str(skill_count)]["icon"] = servant_assist.list[1]
    servant_dict["skill_" + str(skill_count)]["name_sc"] = servant_assist.list[2]
    servant_dict["skill_" + str(skill_count)]["cool_time"] = servant_assist.list[3]
    return servant_dict
```

```python
# 4. 定义技能消耗
def parse_servant_dict_consume(servant_dict, mooncell_text):
    parse_read_flag = 0
    parse_read_string = ""
    parse_read_count = 0
    for line in mooncell_text:
        if line == "": continue
        if line == "===技能强化===":
            parse_read_flag = 1
            servant_dict["skill_consume"] = {}
            continue
        if parse_read_flag >= 1:
            for c in line:
                if c == "{":
                    parse_read_flag = 2
                    parse_read_count += 1
                elif c == "}":
                    parse_read_count -= 1
                if parse_read_flag == 2:
                    parse_read_string += c
                if parse_read_count == 0 and parse_read_flag == 2:
                    parse_read_flag = 3
                    break
            if parse_read_flag == 3:
                break
    servant_assist = mooncell_assist(parse_read_string)
    servant_assist.parse()
    servant_dict["skill_consume"]["consume_rarity"] = int(servant_assist.dict["稀有度"])
    for i in range(1,10):
        servant_dict["skill_consume"]["skill_level_" + str(i)] = {}
        for j in servant_assist.dict[str(i) + "->" + str(i+1)]:
            servant_dict["skill_consume"]["skill_level_" + str(i)][j.list[1]] = int(j.list[2])
    servant_dict["skill_consume"]["total_consume"] = {}
    for j in servant_assist.dict["总计"]:
        servant_dict["skill_consume"]["total_consume"][j.list[1]] = int(j.list[2])
    servant_dict["skill_consume"]
    return servant_dict
```

```python
# 5. 定义灵基再临消耗
def parse_servant_dict_ascension(servant_dict, mooncell_text):
    if servant_dict["basic"]["id"] == 1:
        return servant_dict
    parse_read_flag = 0
    parse_read_string = ""
    parse_read_count = 0
    for line in mooncell_text:
        if line == "": continue
        if line[:7] == "===灵基再临":
            parse_read_flag = 1
            servant_dict["ascension_consume"] = {}
            continue
        if parse_read_flag >= 1:
            for c in line:
                if c == "{":
                    parse_read_flag = 2
                    parse_read_count += 1
                elif c == "}":
                    parse_read_count -= 1
                if parse_read_flag == 2:
                    parse_read_string += c
                if parse_read_count == 0 and parse_read_flag == 2:
                    parse_read_flag = 3
                    break
            if parse_read_flag == 3:
                break
    servant_assist = mooncell_assist(parse_read_string)
    servant_assist.parse()
    servant_dict["ascension_consume"]["consume_rarity"] = int(servant_assist.dict["稀有度"])
    for i in range(0,4):
        servant_dict["ascension_consume"]["ascension_level_" + str(i)] = {}
        for j in servant_assist.dict[str(i) + "->" + str(i+1)]:
            servant_dict["ascension_consume"]["ascension_level_" + str(i)][j.list[1]] = int(j.list[2])
    servant_dict["ascension_consume"]["total_consume"] = {}
    for j in servant_assist.dict["总计"]:
        servant_dict["ascension_consume"]["total_consume"][j.list[1]] = int(j.list[2])
    servant_dict["ascension_consume"]
    return servant_dict
```
